```python
import matplotlib.pyplot as plt
import numpy as np
```

\$ (1) x^2 + x y + y^3 - y = 0\$<br>
\$(2) -y + x^2 + x = \frac{1}{2}\$<br><br>
\$ (1) x = \frac{-y \pm \sqrt{y^2 - 4 (y^3 - y)}}{2}\$<br>
\$ (2) y = x^2 + x - \frac{1}{2}\$<br><br>

```python
Y = np.linspace(2.0,-2.0,100)
X_R = (-Y+(Y**2-4*(Y**3-Y))**0.5)/2
X_L = (-Y-(Y**2-4*(Y**3-Y))**0.5)/2
X_2 = np.linspace(-2.0,2.0,100)
Y_2 = X_2**2 + X_2 -1/2
plt.plot(X_R,Y,X_L,Y,X_2,Y_2)
plt.show()
```

```python
Y = np.linspace(0.5,0.0,100)
X_R = (-Y+(Y**2-4*(Y**3-Y))**0.5)/2
X_L = (-Y-(Y**2-4*(Y**3-Y))**0.5)/2
X_2 = np.linspace(0.0,0.5,100)
Y_2 = X_2**2 + X_2 -1/2
plt.plot(X_R,Y,X_L,Y,X_2,Y_2)
plt.show()
```

\$ (1) x_1 = \frac{-y_1 \pm \sqrt{y_1^2 - 4 (y_1^3 - y_1)}}{2}\$<br>
\$ (2) y_2 = x_2^2 + x_2 - \frac{1}{2}\$<br><br>

\$ D^2 =  \left( x_2 - \frac{-y_1 + \sqrt{y_1^2 - 4 (y_1^3 - y_1)}}{2}\right)^2 + \left( x_2^2 + x_2 + 2  - y_1\right)^2 \$

```python
xlist = np.linspace(-0.5, 0.9, 1000)
ylist = np.linspace(0.0, 1.0, 1000)
X, Y = np.meshgrid(xlist, ylist)
Z = (X - (-Y + (Y-4*(Y**3-Y))**0.5)/2 )**2 + (X**2+X-1/2-Y)**2
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
def D_2(x,y):
    return (x - (-y + (y-4*(y**3-y))**0.5)/2 )**2 + (x**2+x-1/2-y)**2

def dD_2dx(x,y):
    return (D_2(x+0.001,y)-D_2(x,y))/(0.001)

def dD_2dy(x,y):
    return (D_2(x,y+0.001)-D_2(x,y))/(0.001)

x_g = [-0.4]
y_g = [0.9]
for n in range(8*10**2):
    x_g.append(x_g[-1] - 0.001*dD_2dx(x_g[-1],y_g[-1]))
    y_g.append(y_g[-1] - 0.001*dD_2dy(x_g[-1],y_g[-1]))
print(x_g[-1], y_g[-1])
    
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.plot(x_g,y_g,c='r')
plt.show()
```
